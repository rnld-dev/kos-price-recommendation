`Pemecahan data Maps dari Link Maps menjadi informasi Latitude dan Longitude`

```python
import re

# Fungsi untuk mengekstrak latitude dan longitude dari URL
def extract_lat_long(url):
    # Fungsi re.search mencari pola di dalam string url
    match = re.search(r'q=(-?\d+\.\d+),(-?\d+\.\d+)', url)
    if match:
        return float(match.group(1)), float(match.group(2))
    else:
        return None, None

# Menerapkan fungsi ke kolom 'Maps' dan membuat kolom 'Latitude' dan 'Longitude'
df[['Latitude', 'Longitude']] = df['Maps'].apply(lambda x: pd.Series(extract_lat_long(x)))
```

`Menghitung jarak lokasi kos terhadap 9 lokasi kampus (dalam km)`

```python
# Data kampus dan koordinatnya
kampus_data = {
    'UPN': (-7.3327844081176625, 112.7886245733404),
    'UNAIR Kampus Dharmahusada A': (-7.2645080968064235, 112.75858359414417),
    'UNAIR Kampus Dharmawangsa B': (-7.271853056880128, 112.75868463647284),
    'UNAIR Kampus Merr C': (-7.268482055126031, 112.78428250948707),
    'ITS Kampus Sukolilo': (-7.282281496662931, 112.79496820948731),
    'ITS Kampus Manyar': (-7.286417280088548, 112.76375714996585),
    'ITS Kampus Cokroaminoto': (-7.277801244766786, 112.7389771806514),
    'UNESA Kampus Ketintang (Gayungan)': (-7.314968630328221, 112.7271889211307),
    'UNESA Kampus Lidah Wetan (Lakarsantri)': (-7.300743694103851, 112.67271643403186)
}

kampus_data
```

```python
from math import radians, sin, cos, atan2, sqrt

# Fungsi untuk menghitung jarak Haversine
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius bumi dalam kilometer
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])  # Konversi derajat ke radian
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

Tujuan dari baris kode ini adalah untuk menghitung jarak dari setiap lokasi kos-kosan ke setiap kampus yang ada, menggunakan fungsi haversine yang telah kita definisikan sebelumnya. Hasil perhitungan ini akan ditambahkan sebagai kolom baru di DataFrame dengan nama kampus yang bersangkutan.

```python
# Menghitung jarak ke setiap kampus dan menambahkannya ke DataFrame
for kampus, coords in kampus_data.items():
    df[kampus] = df.apply(lambda row: haversine(row['Latitude'], row['Longitude'], coords[0], coords[1]), axis=1)
```

`Membentuk kolom kampus TERdekat dan jaraknya`

```python
# Fungsi untuk menemukan kampus terdekat dan jaraknya
def find_nearest_campus(row):
    distances = row[kampus_data.keys()].astype(float)  # Pastikan semua nilai jarak adalah float
    nearest_campus = distances.idxmin()  # Dapatkan kampus terdekat
    nearest_distance = distances.min()  # Dapatkan jarak ke kampus terdekat
    return pd.Series([nearest_campus, nearest_distance])
```

```python
# Menerapkan fungsi untuk menemukan kampus terdekat dan jaraknya
df[['Kampus Terdekat', 'Jarak ke Kampus Terdekat']] = df.apply(find_nearest_campus, axis=1)
```
